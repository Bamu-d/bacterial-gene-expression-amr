```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
```

```python
#deduce genes differentially expressed when compared to k12 
#thresholds padj <= 0.0001, LFC >=2 


path = '/Users/bamudamaris/Library/CloudStorage/GoogleDrive-fufordamaris@gmail.com/My Drive/Data/DEG/DESeq2_output/'

full_table = None

for strain in os.listdir(path):
    if strain.startswith('NT'):
        df = pd.read_csv(path + strain)
        # filter data set for those with significant DEGs
        df = df[(df['log2FoldChange'].abs() >= 2) & (df['padj'] <= 0.0001)]
        # rename columns 
        df = df.rename(columns={'Unnamed: 0':'gene_id', 'log2FoldChange': strain[:-4]})
        # drop all irrelevant columns 
        df = df.set_index('gene_id').drop(columns=['baseMean', 'lfcSE', 'stat', 'pvalue', 'padj'], axis=1)
        # loop to merge multiple dataframes
        if full_table is None:
            full_table = df
        else:
            full_table = pd.merge(full_table, df, left_index=True, right_index=True, how='outer')

full_table.index = ["_".join(i.split("_cds_")[1].split("_")[:-1]) for i in full_table.index]            

# Count the occurrences of each gene being differentially expressed across strains
full_table.loc['Total_DEGs'] = full_table.notnull().sum()


full_table
```

```python
# Extract the Total_DEGs per strain
degs_count = full_table.loc['Total_DEGs'].reset_index()
degs_count.columns = ['Strain', 'Total_DEGs']
degs_count = degs_count.set_index('Strain')
degs_count

```

```python
#compute top variable genes across strains. 

# using vst normalised count. 
top = pd.read_csv('/Users/bamudamaris/Library/CloudStorage/GoogleDrive-fufordamaris@gmail.com/My Drive/Data/Datasets/vst_batch_corrected.tsv', sep = ',',index_col = 0)
#top.columns = [x.split('_')[0] for x in top.columns]
top.index = ["_".join(i.split("_cds_")[1].split("_")[:-1]) for i in top.index]
top

```

```python
# Extract base sample names without the replicate suffix
base_sample_names = top.columns.str.rsplit('_', n=1).str[0]

# Average duplicate samples
averaged_exp = top.groupby(base_sample_names, axis=1).mean()

#compute standard deviation for each gene across the samples 
averaged_exp['std_dev'] = averaged_exp.std(axis=1)

# Sort by variability and select genes
top_var_genes = averaged_exp.sort_values(by='std_dev', ascending=False).head(50)
top_var_genes
```

```python
common_elements = list(set(full_table.index) & set(top_var_genes.index))

top50 = top_var_genes[top_var_genes.index.isin(common_elements)].T.drop('std_dev')
```

```python
genes = pd.read_csv('/Users/bamudamaris/Library/CloudStorage/GoogleDrive-fufordamaris@gmail.com/My Drive/Data/Datasets/gene_positions.txt', sep = '\t', index_col = 1)
genes
```

```python
top50 = top50.T
top50= pd.merge(top50, genes['gene_name'], left_index = True, right_index = True, how = 'left')
top50
#top50.to_csv('top_var_50.tsv', sep = '\t')
```
